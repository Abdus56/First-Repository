# Missing Values

```python
df.isnull().sum()   ## this will give us sum of null values of all columns in dataset
```

```python
## Drop null values and edit it in dataset
## this will drop all the rows with null values not only columns of specific missing values inplace will also allows to edit the dataset.
## if you don't use inpace then when again call df dataset it'll those missing values too.
df.dropna(inplace = True)    
```

```python
df
```

```python
## Fill null values:

df_updated = df.fillna(0)
```

```python
df_updated.horsepower[df_updated.horsepower == 0] ## we get the values that we fill with 0.
```

```python
df.fillna(method = "bfill")  ## this will fill the missing vaues by values that are next to that values. Backward fill.
```

```python
## Interpolate function fill the missing values by drwaing a linear line between values and fill the missing values.
df.interpolate()
```

```python
df.interpolate(method = 'cubic')   ## in that case method is cubic so it will take power of 3 and then fill missing values.
```
